```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from scipy.spatial.distance import cosine as cosine
from sklearn.metrics import accuracy_score
```

```python
data = np.arange(0,600, dtype=float)-np.random.random(600)*75
```

```python
data = data.reshape((200,3))
```

```python
scaler = StandardScaler()
data_t = scaler.fit_transform(data)
data_t
```

```python
pca = PCA()
pca.fit(data_t)
pca.explained_variance_ratio_
```

## KNN Regressor with Cosine

```python
knn_model = KNeighborsRegressor(metric=cosine)
```

```python
np.random.shuffle(data_t)
```

```python
X = data_t[:,0:-1]
y = data_t[:,-1]
```

```python
t = len(X)/5*4
X_train = X[:t]
y_train = y[:t]
X_test = X[t:]
y_test = y[t:]
```

```python
knn_model.fit(X_train, y_train)
```

```python
y_pred = knn_model.predict(X_test)
```

```python
knn_model.score(X_test, y_test)
```
